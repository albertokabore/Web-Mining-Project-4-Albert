```python
import requests
import json
from spacytextblob.spacytextblob import SpacyTextBlob
import spacy

# Step 1: Save the lyrics to a JSON file (run this step only once if the file doesn't exist)
result = json.loads(requests.get('https://api.lyrics.ovh/v1/They Might Be Giants/Birdhouse in your soul').text)
filename = 'lyrics.json'
with open(filename, 'w') as json_file:
    json.dump(result, json_file)

# Step 2: Read the contents of the JSON file
with open(filename, 'r') as json_file:
    data = json.load(json_file)

# Extract the lyrics
lyrics = data.get("lyrics", "")
print("Lyrics of the song:\n", lyrics)

# Step 3: Perform sentiment analysis using spaCyTextBlob
# Load spaCy and add spaCyTextBlob pipeline
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe('spacytextblob')

# Analyze the lyrics
doc = nlp(lyrics)
polarity = doc._.polarity

# Step 4: Print the polarity score
print(f"\nPolarity score: {polarity}")

# Step 5: Interpret the polarity score
# Comment: The polarity score ranges from [-1.0, 1.0].
# A score > 0 indicates positive sentiment, < 0 indicates negative sentiment, and 0 indicates neutrality.
if polarity > 0:
    print("The lyrics have a more positive connotation.")
elif polarity < 0:
    print("The lyrics have a more negative connotation.")
else:
    print("The lyrics have a neutral connotation.")
```
